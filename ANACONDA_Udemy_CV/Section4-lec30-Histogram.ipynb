```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
horse = cv.imread("DATA/horse.jpg")
horse1 = cv.cvtColor(horse,cv.COLOR_BGR2RGB)

rainbow = cv.imread("DATA/rainbow.jpg")
rainbow1 = cv.cvtColor(rainbow,cv.COLOR_BGR2RGB)

brick = cv.imread("DATA/bricks.jpg")
brick1 = cv.cvtColor(brick,cv.COLOR_BGR2RGB)

plt.imshow(horse1)
```

```python
# Making histogram 

hist1 = cv.calcHist([horse],channels=[1],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist1)
```

```python
# calculating histrogram for all three colors

bricks2 = brick

color = ['b','g','r']
for i,col in enumerate(color):
    hist2 = cv.calcHist([bricks2],[i],None,[256],[0,256])
    plt.plot(hist2,color = col)
    plt.xlim([0,256])

plt.title("HISTOGRAM OF BRICKS")
```

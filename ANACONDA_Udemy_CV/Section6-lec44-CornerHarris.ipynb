```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img1 = cv.imread("DATA/flat_chessboard.png")
img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
```

```python
img1gray = cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
plt.imshow(img1gray, cmap = 'gray')
```

```python
floatgray = np.float32(img1gray)
img1gray
hc = cv.cornerHarris(floatgray,2,3,0.04)
plt.imshow(hc,cmap='gray')
```

```python
hc = cv.dilate(hc,None)

img1[hc > 0.01*hc.max()] = [255,0,0]
plt.imshow(img1)
```

```python
img2 = cv.imread("DATA/real_chessboard.jpg")
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
plt.imshow(img2)

img2gray = cv.cvtColor(img2,cv.COLOR_RGB2GRAY)
plt.imshow(img2gray, cmap='gray')
```

```python
img2float = np.float32(img2gray)

hc2 = cv.cornerHarris(img2float,2,3,0.04)
hc22 = cv.dilate(hc2,None)

img2[hc22 > 0.01*hc22.max()] = [255,0,0]

plt.imshow(img2,cmap='gray')
```

```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
# reading 3 images in Grayscale

img1 = cv.imread("DATA/Denis_Mukwege.jpg", 0)
img2 = cv.imread("DATA/Nadia_Murad.jpg", 0)
img3 = cv.imread("DATA/solvay_conference.jpg", 0)
```

```python
# getting cassifier to classify face features 

face_cascade = cv.CascadeClassifier("DATA/haarcascades/haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img = img.copy()
    
# detectmultiscale function takes an image to dtetct the faces or anyother features based on the cascade and 
# return back the x,y,width,height, to draw circle around the feature detected
    face_rects = face_cascade.detectMultiScale(face_img,)
    
# lets draw rectangle 

    for x,y,w,h in face_rects:
        cv.rectangle(face_img,(x,y), (x+w,y+h), (255,0,0),5)
    
    return face_img
```

```python
def display(img):
    fig = plt.figure(figsize=(10,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display(detect_face(img3))
```

```python
eye_cascade = cv.CascadeClassifier("DATA/haarcascades/haarcascade_eye.xml")
```

```python
def detect_eyes(img):
    eye_img = img.copy()
    
# detectmultiscale function takes an image to dtetct the faces or anyother features based on the cascade and 
# return back the x,y,width,height, to draw circle around the feature detected
    eye_rects = eye_cascade.detectMultiScale(eye_img,scaleFactor =1.2)
    
# lets draw rectangle 

    for x,y,w,h in eye_rects:
        cv.rectangle(eye_img,(x,y), (x+w,y+h), (255,0,0),10)
    
    return eye_img
```

```python
display(detect_eyes(img3))
```

```python
cap = cv.VideoCapture(0)

while True: 
    ret, frame = cap.read(0)
    
    frame = detect_face(frame)
    
    cv.imshow("video", frame)
    
    k = cv.waitKey(1) 
    if k == 27:
        break
cap.release()     
cv.destroyAllWindows()
```

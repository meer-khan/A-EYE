```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread("DATA/rainbow.jpg", 0)
# img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
# imggray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
# THresholding

ret , th1 = cv.threshold(img, 127 ,255 , cv.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(th1, cmap='gray')
```

```python
# THRESHOLDING BINARY INVERSE

ret1 , th2 = cv.threshold(img, 127 ,255 , cv.THRESH_BINARY_INV)

plt.imshow(th2 , cmap='gray')
```

```python
#  THRESHOLD TRUCATION


ret2 , th3 = cv.threshold(img, 127 ,255 , cv.THRESH_TRUNC)

plt.imshow(th3 , cmap='gray')
```

```python
# READING ANOTHER IMAGE 

img2 = cv.imread("DATA/crossword.jpg", 0)
plt.imshow(img2, cmap='gray')
```

```python
def display(img):
    fig = plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display(img2)
```

```python
# APPLYING THRESHOLDING

ret3, th4 = cv.threshold(img2,190,255,cv.THRESH_BINARY)
display(th4)
```

```python
# Adaptive threshold , that calculate threshold + smoothing the image using GAUSSIAN AND MEDIANBLUR FILTERS

adth1 = cv.adaptiveThreshold(img2,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,8 )
display(adth1)
```

```python
# blending

blended = th4+adth1

blended2 = cv.addWeighted(th4,0.8,adth1,0.5,0)

display(blended2)
```

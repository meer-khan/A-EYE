```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
def load():
    img = cv.imread("DATA/sudoku.jpg",0)
    return img
    
```

```python
def display(img1):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img1, cmap='gray')
```

```python
img = load()
display(img)
```

```python
# Applying sobel filter in Y direction

sobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=5,)
display(sobelx)
```

```python
# Applying sobel filter in X direction

sobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=5,)
display(sobely)
```

```python
sobelxy = sobelx + sobely
display(sobelxy)
```

```python
laplacianimg = load()
display(laplacianimg)
```

```python
laplacian = cv.Laplacian(laplacianimg, cv.CV_64F)
display(laplacian)
```

```python
newimg = laplacian + sobelxy
display(newimg)
```

```python
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
img1 = cv.imread("DATA/flat_chessboard.png")
img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)

plt.imshow(img1)
#  one thing to keep in mind that image should be look like a chess board to make the track of the object possible
# at any direction
# grid should be specifically made like a chessboard
```

```python
# FINDING CHESS BOARD CORNERS
found , corners = cv.findChessboardCorners(img1,(7,7))

# found is a boolean to check if this method has found any pattern or not 
#  corners gives numpy array of cordinates where it find corners 

# We will mark the corners by using method below 

img2 = cv.drawChessboardCorners(img1,(7,7),corners,found)
```

```python
plt.imshow(img2)
```

```python
img3 = cv.imread("DATA/dot_grid.png")

plt.imshow(img3)
```

```python
# NOW WE WILL DO CONTOUR DETECTION: 
# Reading image in grayscale in necessary
img5 = cv.imread("DATA/internal_external.png",0)
print(img5[500,150])
plt.imshow(img5 , cmap='gray')


```

```python
# Applying contour function

img6, contours, hirerchy = cv.findContours(img5,cv.RETR_CCOMP,cv.CHAIN_APPROX_SIMPLE)
hirerchy
```

```python
#  Now we can contours points of img6 
#  and now we can plot them on our own image 
#  so we will create a black image using no.zeros

external_contours = np.zeros(img6.shape)
external_contours
len(contours)
```

```python
# Now we will draw contours using contours and herirarchy

for i in range(len(contours)):
    if hirerchy[0][i][3] != -2:
        cv.drawContours(external_contours,contours,i,(255,0,0),10)
    
```

```python
plt.imshow(external_contours, cmap = 'gray')
```

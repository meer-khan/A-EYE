```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread("DATA/car_plate.jpg")
```

```python
def display(img):
    img1 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    fig = plt.figure(figsize = (10,12))
    ax = fig.add_subplot(111)
    ax.imshow(img1)
    return img1
```

```python
display(img)
```

```python
number_plate = cv.CascadeClassifier("DATA/haarcascades/haarcascade_russian_plate_number.xml")
```

```python
def numberplate(img):
    
    plate_rects = number_plate.detectMultiScale(img, scaleFactor=1.2)
    img_copy = img.copy()
    
    
    for x,y,w,h in plate_rects:
        
        img_rectangle = cv.rectangle(img_copy, (x,y),(x+w,y+h), (0,0,255),2 )
    return img_copy
```

```python
dogrgb = cv.cvtColor(dog , cv.COLOR_BGR2RGB)
dogrgbcopy = cv.cvtColor(dog , cv.COLOR_BGR2RGB)
plt.imshow(dogrgbcopy)
plt.imshow(dogrgb)
```

```python
flippedDog=cv.flip(dogrgb,0)
plt.imshow(flippedDog)
```

```python
cv.rectangle(dogrgb , pt1 = (250,350), pt2=(600,750), color = (255,0,0), thickness = 7,)
plt.imshow(dogrgb)
```

```python
doggie = cv.cvtColor(dog,cv.COLOR_BGR2RGB)
plt.imshow(doggie)

def mou(event,x,y,flags,params):
    if event==cv.EVENT_RBUTTONUP:
        cv.circle(dog,(x,y),radius = 40 , color=(255,0,0),thickness=10)

cv.namedWindow(winname='imgee')
cv.setMouseCallback('imgee',mou)
while True:
    cv.imshow("imgee",dog)
    if cv.waitKey(1) & 0xFF == 27:
        break


cv.destroyAllWindows()
```
